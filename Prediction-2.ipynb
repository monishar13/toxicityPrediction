```python
import pandas as pd
import numpy as np 
```

```python
train_data_desc = pd.read_csv("./df_with_200_descriptors_csv.csv")
test_data_desc = pd.read_csv("./df_test.csv")
```

```python
train_data_desc.describe()
```

```python
test_data_desc
```

```python
# Dropping null values in training data 
train_data_desc = train_data_desc.dropna()
# Replacing null value with 0 in test data 
test_data_desc = test_data_desc.fillna(0)
```

```python
X = train_data_desc.drop("expected",axis=1)
y = train_data_desc["expected"]
```

```python
# Using random forest algorithm 
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(random_state=13)
```

```python
model.fit(X,y)
```

```python
predict = model.predict(test_data_desc)
```

```python
print(np.count_nonzero(predict==1))
print(np.count_nonzero(predict==2))
```

```python
submission_file = pd.read_csv("./sample_submission.csv")
```

```python
submission_file.Predicted = predict
```
